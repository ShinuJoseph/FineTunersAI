```python
# import data

PATH = '..\..\data\cleaned_output.csv'

df = pd.read_csv(PATH)
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Assuming your original dataframe is named 'df'
# df = pd.read_csv('your_file.csv') # Uncomment if reading from a CSV file

# Function to check if a paragraph is likely an ad or irrelevant
def is_irrelevant_paragraph(paragraph):
    text = paragraph.get_text().strip()
    if not text:
        return True
    if "menu-section-heading" in paragraph.get("class", []):
        return True
    if text.startswith("©") or "byline-bio" in paragraph.get("class", []):
        return True
    if paragraph.find('a') and 'href' in str(paragraph):
        return True
    return False

# Function to process HTML content and extract relevant paragraphs
def process_html(content):
    soup = BeautifulSoup(content, 'html.parser')
    paragraphs = soup.find_all('p')
    
    # Filter out irrelevant paragraphs
    filtered_paragraphs = [p.get_text().strip() for p in paragraphs if not is_irrelevant_paragraph(p)]
    
    return filtered_paragraphs

# Creating a new dataframe for cleaned data
cleaned_data = []

for index, row in df.iterrows():
    paragraphs = process_html(row["content"])
    for i, para in enumerate(paragraphs):
        cleaned_data.append({
            "link": row["link"],
            "title": row["title"],
            "paragraph_number": i + 1,
            "content": para
        })

cleaned_df = pd.DataFrame(cleaned_data)

# Output the cleaned dataframe
display(cleaned_df.head(20))  # Prints first few rows of the dataframe


# cleaned_df.to_csv('cleaned_data.csv', index=False) # Uncomment to save to a CSV file

```
